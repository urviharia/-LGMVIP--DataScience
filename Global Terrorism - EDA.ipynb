### Cleaning the data

```python
terror=terror[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude","longitude","location",
               "summary","attacktype1_txt","targtype1_txt","gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
terror.head()
```

```python
terror.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region",
                       "provstate":"Province/State","city":"City","latitude":"Latitude","longitude":"Longitude",
                       "location":"Location","summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                       "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type","nkill":"Killed",
                       "nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
terror.head()
```

```python
terror.info()
```

```python
terror.shape
```

```python
terror.isnull().sum()
```

```python
terror["Killed"]=terror["Killed"].fillna(0)
terror["Wounded"]=terror["Wounded"].fillna(0)
terror["Casualty"]=terror["Killed"]+terror["Wounded"]
```

```python
terror.describe()
```

## Visualizing the data

### 1 . Year wise attacks

##### A . Number of attacks in each year

```python
attacks=terror["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind="bar",figsize=(20,10),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

Most number of attacks(16903) are in 2014

Least number of attacks(471) are in 1971

##### B . Total Casualties (Killed + Wounded) in each Year

```python
tc=terror[["Year","Casualty"]].groupby("Year").sum()
tc.head()
```

```python
tc.plot(kind="bar",color="forestgreen",figsize=(20,10))
plt.title("Year wise Casualties",fontsize=13)
plt.xlabel("Years",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

```python
print("Maximum casualty is " , tc["Casualty"].max(),  "for the year", tc[tc["Casualty"]==85618.0].index.values) 
print("Minimum casualty is " , tc["Casualty"].min(),  "for the year", tc[tc["Casualty"]==255.0].index.values) 
```

### 2. Region wise attacks

##### A . Distribution of Terrorist Attacks over Regions from 1970-2017

```python
reg=pd.crosstab(terror.Year,terror.Region,rownames=["Year"],colnames=["Region"])
reg.head() 
```

```python
reg.plot(kind="area",stacked=False, alpha=0.9,figsize=(20,20))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

##### B.Total Terrorist Attacks in each Region from 1970-2017

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
rt=regt["Total"].sort_values(ascending=False)
rt
```

```python
rt.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

##### C.Total Casualties (Killed + Wounded) in each Region

```python
Tc=terror[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
Tc
```

```python
Tc.plot(kind="bar",figsize=(15,6))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

##### D.Killed in each Region

```python
Kr=terror[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed",ascending=False)
Kr
```

```python
Kr.plot(kind="bar",color="red",figsize=(15,6))
plt.title("People Killed in each Region")
plt.xlabel("Regions")
plt.ylabel("Number of People Killed")
plt.show()
```

##### E . Wounded in each Region

```python
Wr=terror[["Region","Wounded"]].groupby("Region").sum().sort_values(by="Wounded",ascending=False)
Wr
```

```python
Wr.plot(kind="bar",color="purple",figsize=(15,6))
plt.title("People Wounded in each Region")
plt.xlabel("Regions")
plt.ylabel("Number of People Wounded")
plt.show()
```
